```python
from colour import Color

red = Color("red")
colors = list(red.range_to(Color("green"), 21))

def convert_to_rgb(minval, maxval, val):
    if math.isnan(val):
        return red.hex

    t = (val - minval) / (maxval - minval) #[0..1]
    c = colors[int(t * 20 )]
    return c.hex
```

```python
import pandas as pd
df = pd.read_csv('data-20190606T1257-structure-20190606T1257.csv')
```

```python
# print(df[25:])
```

```python
data = [go.Bar(x=df.iloc[1:,0], y=df.iloc[1:,1])]
iplot(data, filename='basic-bar')
```

```python
df = df.drop(df.index[[0, 1, 19, 45, 60, 67, 80]])
df.dropna()
data = [go.Bar(x=df.iloc[:,0], y=df.iloc[:,1])]
iplot(data, filename='basic-bar')
```

```python
a, b = min(df.iloc[:,1]), max(df.iloc[:,1])
df['name'] = df.iloc[:,0]
df['value'] = df['план за 2018 г, га'].map(lambda t: convert_to_rgb(a, b, t) )
df = df.dropna()
```

```python
json_data = {}
for index, row in df.iterrows():
    json_data[row['name']] = row['value']
```

```python
# print(json_data)
with open('heatmap.json', 'w') as outfile:
    json.dump(json_data, outfile)
```
