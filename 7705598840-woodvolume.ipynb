```python
from colour import Color

red = Color("seagreen")
colors = list(red.range_to(Color("lightsalmon"), 21))

def convert_to_rgb(minval, maxval, val):
    if math.isnan(val):
        return red.hex

    t = (val - minval) / (maxval - minval) #[0..1]
    c = colors[int(t * 20 )]
    return c.hex
```

```python
import pandas as pd
df = pd.read_csv('data-20190605T1013-structure-20171226T0931.csv', thousands=r',')
df['The volume of harvested wood'] = df['The volume of harvested wood'].apply(lambda t: float(t.replace(".","").replace(",",".").replace(" ", "")))
df['Subject'] = df['Subject'].apply(lambda t: t.replace("обл","область"))
```

```python
data = [go.Bar(x=df.iloc[1:,0], y=df.iloc[1:,1])]
iplot(data, filename='basic-bar')
```

```python
df = df.drop(df.index[[0, 1, 19, 45, 60, 67, 80]])
df.dropna()

data = [go.Bar(x=df.iloc[:,0], y=df.iloc[:,1])]
iplot(data, filename='basic-bar')
print(df)
```

```python
a, b = float(min(df.iloc[:,1])), float(max(df.iloc[:,1]))
df['name'] = df.iloc[:,0]
df['value'] = df.iloc[:,1]
df['value'] = df['value'].map(lambda t: convert_to_rgb(a, b, t) )
```

```python
json_data = {}
for index, row in df.iterrows():
    json_data[row['name']] = row['value']
```

### Save data to display

```python
with open('7705598840-woodvolume_map.json', 'w') as outfile:
    json.dump(json_data, outfile)
```
